### Solid phase

```python
nsteps_s = 2000 # steps per block
nblocks_s = 20 # blocks

epot_s_data = np.loadtxt("epot_solid.out")
ekin_s_data = np.loadtxt("ekin_solid.out")
etot_s_data = np.loadtxt("etot_solid.out")
temp_s_data = np.loadtxt("temp_solid.out")
press_s_data = np.loadtxt("press_solid.out")

epot_s_blocks = epot_s_data[:, 0]
ekin_s_blocks = ekin_s_data[:, 0]
etot_s_blocks = etot_s_data[:, 0]
temp_s_blocks = temp_s_data[:, 0]
press_s_blocks = press_s_data[:, 0]

epot_s_av = epot_s_data[:, 2]
ekin_s_av = ekin_s_data[:, 2]
etot_s_av = etot_s_data[:, 2]
temp_s_av = temp_s_data[:, 2]
press_s_av = press_s_data[:, 2]

epot_s_err = epot_s_data[:, 3]
ekin_s_err = ekin_s_data[:, 3]
etot_s_err = etot_s_data[:, 3]
temp_s_err = temp_s_data[:, 3]
press_s_err = press_s_data[:, 3]
```

```python
# potential energy
plt.errorbar(epot_s_blocks*nsteps_s, epot_s_av, yerr=epot_s_err)
plt.xlabel('#steps')
plt.ylabel('potential energy')
plt.title('Potential energy')
plt.grid(True)
plt.show()

# kinetic energy
plt.errorbar(ekin_s_blocks*nsteps_s, ekin_s_av, yerr=ekin_s_err)
plt.xlabel('#steps')
plt.ylabel('kinetic energy')
plt.title('Kinetic energy')
plt.grid(True)
plt.show()

# total energy
plt.errorbar(etot_s_blocks*nsteps_s, etot_s_av, yerr=etot_s_err)
plt.xlabel('#steps')
plt.ylabel('total energy')
plt.title('Total energy')
plt.grid(True)
plt.show()

# temperature
plt.errorbar(temp_s_blocks*nsteps_s, temp_s_av, yerr=temp_s_err)
plt.axhline(y=0.8, color='r', linestyle='--', label='target temperature')
plt.xlabel('#steps')
plt.ylabel('temperature')
plt.title('Temperature')
plt.grid(True)
plt.legend()
plt.show()

# pressure
plt.errorbar(press_s_blocks*nsteps_s, press_s_av, yerr=press_s_err)
plt.xlabel('#steps')
plt.ylabel('pressure')
plt.title('Pressure')
plt.grid(True)
plt.show()
```

### Comments

MD simulations are based upon long term (total) energy conserving algorithms, such as Verlet's algorithm (notice the difference in scale between the plots of total anergy and kinetic and potential energy). As a consequence, the initial temperature of the simultion is not conserved in general. In fact, if the initial configuration is significantly different from the typical equilibrium configuration, we expect to see variations of kinetic and potential energy as the system approaches equilibrium. Since $T$ is proportional to the average kinetic energy per particle, it will drift away from the initial value in the first part of the simulation. For all of the simulations performed here, the initial configuration was an fcc cristalline structure, which is very different from the typical equilibrium configuration at least for the liquid and the gas phase. Solid Ar has indeed a fcc structure, albeit with a lattice spacing ($1.09 \sigma$) different from the one encoded in the simulation. For these reasons the input temperatures are different from the target temperatures of the simulation. Moreover, the simulations were run for a certain number of equilibration steps (namely 10000 for the solid phase and 110000 for the liquid and the gas phase) before taking measurements. The number of equilibration steps was chosen based on the plot of potential energy. </p>

The simulations of the liquid and the solid phase consist of 20 blocks of length (i. e. number of steps) 2000, the simulation of the gas phase consists of 20 block of length 5000 instead, because the observables displayed greater fluctuations in this case.
